```python
df = pd.read_csv("/content/drive/MyDrive/DataAnalysis/data/bestsellers.csv")
```

```python
df.head()
```

```python
#Find the books written by Pete Souza
df[df["Author"] == "Pete Souza"]
```

```python
# Find the books that are under 10 dollars
df[df["Price"] < 10]
```

```python
#Find the books that have a price between 50 and 60 dollars
df[df["Price"].between(50, 60)]
```

```python
#Find all the books written by Kristin Hannah, Andy Weir, or Delia Owens
authors = ["Kristin Hannah", "Andy Weir", "Delia Owens"]
df[df["Author"].isin(authors)]

#df[(df["Author"] == "Kristin Hannah") | (df["Author"] == "Andy Weir") | (df["Author"] == "Delia Owens")]
```

```python
df.info()
```

```python
#Find the Non Fiction books that are rated 4.9
genre = df["Genre"] == "Non Fiction"
rating = df["User Rating"] == 4.9
df[genre & rating]
```

```python
#Find the fiction book with the lowest User Rating
# genre = df["Genre"] == "Fiction"
# lowest = df["User Rating"].min()
# lowest_rating = df["User Rating"] == lowest
# df[genre & lowest_rating]

df[df["Genre"] == "Fiction"].sort_values("User Rating")
```

```python
#Find 2012's top 5 Fiction books with the most Reviews
df_new = df[(df["Year"] == 2012) & (df["Genre"] == "Fiction")]
df_new.sort_values("Reviews", ascending=False).head()
```

```python
#Create a bar plot showing the 5 authors who have the most books with a rating under 4.5
low_rated = df[df["User Rating"] < 4.5]
#df[low_rated].Author.value_counts().head().plot(kind="bar")
low_rated["Author"].value_counts().head().plot(kind="barh")
```
